```python
import glob
```

```python
def update_notebook(notebook_content):
    cells = notebook_content['cells']
    new_cells = []
    idx = 0
    for cell in cells:
        if cell.get('outputs'):
            cell['outputs'] = []
        if cell['cell_type'] == 'code':
            cell['execution_count'] = None
        if idx == len(cells):
            break
        if len(cell['source']) > 0 and not cell['source'][0].startswith('%%HTML'):
            new_cells.append(cells[idx])
        if len(cell['source']) > 0 and cell['source'][0].startswith('%%HTML') and cells[idx - 1]['cell_type'] == 'markdown' and cells[idx + 1]['cell_type'] == 'markdown':
            cells[idx - 1]['source'] = cells[idx - 1]['source'] + cells[idx + 1]['source']
            idx += 1
        idx += 1
    notebook_content['cells'] = new_cells
    return notebook_content
```

```python
notebooks = glob.glob('python-and-sql-material/*/*.ipynb')
```

```python
import json
```

```python
import subprocess
for notebook in notebooks:
    try:
        print(f'processing {notebook}')
        notebook_content = json.load(open(notebook))
        notebook_content_new = update_notebook(notebook_content)
        notebook_file = open(notebook, 'w')
        json.dump(notebook_content_new, notebook_file, indent=2)
        notebook_file.close()
        subprocess.check_call(f'jupyter trust "{notebook}"', shell=True)
    except:
        print(f'failed {notebook}')
```
